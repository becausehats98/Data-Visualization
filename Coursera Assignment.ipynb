```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('''Topic_Survey_Assignment.csv''', index_col = 0)
df
```

```python
df = df.sort_values(by = '''Very interested''', ascending = False)
df
```

```python
total_respondents = 2233
df = round(df/total_respondents*100, 2)
df
```

```python
ax = df.plot.bar(figsize = (20, 8), width = 0.8, color = ['#5cb85c', '#5bc0de', '#d9534f'])
ax.set_title('''Percentage of Respondent's Intereset in Data Science Areas''', fontsize = 16)
ax.set_xlabel('''Topic''', fontsize = 14)
ax.set_ylabel('''Percentage''', fontsize = 14)
ax.tick_params(axis='both', labelsize=14)
for topic in range(len(df)):
        ax.annotate(df.iloc[topic,0], (topic - 0.38, df.iloc[topic, 0] + 1), fontsize = 14)
        ax.annotate(df.iloc[topic,1], (topic - 0.11, df.iloc[topic, 1] + 1), fontsize = 14)
        ax.annotate(df.iloc[topic,2], (topic + 0.18, df.iloc[topic, 2] + 1), fontsize = 14)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(labels = df.columns, loc = 'upper right', fontsize = 14)
plt.show()
```
